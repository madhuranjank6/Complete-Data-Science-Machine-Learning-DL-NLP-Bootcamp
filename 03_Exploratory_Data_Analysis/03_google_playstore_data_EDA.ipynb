### EDA And Feature Engineering of Google Play Store Dataset
1. Problem Statement: Today, 1.85 million different apps are available for users to download. Android users have even more from which to choose, with 2.5 million available through the Google Play Store. These apps have come to play a huge role in the way we live our lives today. Our objective is to find the most popular category, find the apps with the largest number of installs, the app with largest size, etc

```python
## Dropping the row, which is not convertiable in numeric
df_copy = df.copy()
```

```python
df_copy = df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
## Converting the reviews datatype to int
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df['Size'].isnull().sum()
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('M','000')
df_copy['Size'] = df_copy['Size'].str.replace('k','')
df_copy['Size'] = df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Size']
```

```python
df.head()
```

```python
df['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
chars_to_remove = ['+',',','$']
columns_to_clean = ['Installs','Price']
for item in chars_to_remove:
    for cols in columns_to_clean:
        df_copy[cols] = df_copy[cols].str.replace(item,'') 
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
## Handling the last update feature
df_copy["Last Updated"].unique()
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
df_copy['Day'] = df_copy['Last Updated'].dt.day
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['year'] = df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
## EDA
df_copy.info()
```

```python
df_copy[df_copy.duplicated('App')].shape
```

```python
### There are duplicate records available in the dataset
## Dropping all the duplicates
df_copy = df_copy.drop_duplicates(subset=['App'],keep='first')
```

```python
df_copy.shape
```

```python
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

print(f'we have {len(numeric_features)} Numerical features: {numeric_features}')
print(f'we have {len(categorical_features)} Categorical features: {categorical_features}')
```

```python
## Proportion of count data on categorical columns
for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print("--------------------------")
```

```python
## Visualization
plt.figure(figsize=(15,15))
plt.suptitle("Univariate analysis of Numerical Feature",fontsize=20,fontweight='bold', alpha=0.8,y=1.)
for i in range(0,len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[numeric_features[i]],shade=True,color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

### Observations
- Rating and Year is left skewed while Reviews,Size, Installs and Price are Right skewed

```python
plt.figure(figsize=(15,15))
plt.suptitle("Univariate analysis of Categorical Feature",fontsize=20,fontweight='bold', alpha=0.8,y=1.)
category = ['Type','Content Rating']
for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df_copy[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
